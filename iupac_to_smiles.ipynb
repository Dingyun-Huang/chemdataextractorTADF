#### Convert IUPAC names into SMILES strings

This is achieved with the Open Parser for Systematic IUPAC nomenclature ([OPSIN](https://opsin.ch.cam.ac.uk/)) module and the original paper is [here](https://pubs.acs.org/doi/full/10.1021/). The python wrapper for the JAVA package is provided.

```python
from pyopsin import pyopsin
```

```python
# an example record resulted from ChemDataExtractor
record = {"Compound": {"names": ['2-(9H-Carbazol-9-yl)thianthrene '
                                    '5,5,10,10-tetraoxide',
                                 '2-(9H-carbazol-9-yl)thianthrene '
                                    '5,5,10,10-tetraoxide',
                                 'CZ-TTR']}}
# Change the path to the actual path of the OPSIN .jar file on your system
path = "E:/PhD/chemdataextractorTADF/pyopsin/opsin_cli.jar"
nts = pyopsin.NameToSmiles(path)
smiles = str(nts.name_to_smiles(record["Compound"]["names"][0]))
print(smiles)
```

#### Use RDKit to visualize the molecule

```python
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
molecule = Chem.MolFromSmiles(smiles)
```

```python
Draw.MolToImage(molecule)
```
